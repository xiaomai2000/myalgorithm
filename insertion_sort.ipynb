```python
def insertion_sort(arr, simulation=False):
    
    iteration = 0
    if simulation:
        print("iteration",iteration,":",*arr)
        
    for i in range(len(arr)):
        cursor = arr[i]
        pos = i
        print("Before iteration: ", *arr)
        print("     ---------> i: ", i, " cursor :", cursor , "   pos=", pos)
        
        while pos > 0 and arr[pos - 1] > cursor:
            # Swap  the number down the list
            arr[pos] = arr[pos - 1]
            pos = pos - 1
        # Break and do the final swap
        arr[pos] = cursor
        
        print("After iteration: ", *arr)
        print("     ---------> i: ", i, " cursor :", cursor , "   pos=", pos, "\n")
        
        if simulation:
                 iteration = iteration + 1
        

    return 
```

```python
tarr = [2,8,5,3,9]
insertion_sort(tarr, True)
```
