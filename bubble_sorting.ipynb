```python
def bubble_sort(arr):
    print("Going to sort: {0}".format(arr))
    
    #Swap two number in arr[]
    def swap(i, j):
        arr[i], arr[j] = arr[j], arr[i]
        
    
    # Start sorting
    nlen = len(arr)
    swapped = True
    x = -1
    
    while (swapped and (x < nlen -2)) :
        x = x + 1
        swapped = False
        
        for i in range(1, nlen - x):
            if arr[i-1] > arr[i]:
                # If greater than, swap;
                swap(i-1, i)
                swapped = True
    
        print("After round {0}, sorted array is: {1}".format(x, arr))
    
    
    return arr
    
```

```python
arr = [1,5,8,12,6,9]
bubble_sort(arr)
```

```python
arr = [19,15,8,7,6,5]
bubble_sort(arr)
```

```python
arr = [19,15,8,7,6,5,-2]
bubble_sort(arr)
```
